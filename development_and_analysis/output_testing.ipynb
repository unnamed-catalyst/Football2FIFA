```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

```python
video_path = '../input_videos/08fd33_4.mp4'
cap = cv2.VideoCapture(video_path)
ret, frame = cap.read()
```

```python
plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB
plt.axis('off')
plt.show()
```

```python
# Get the width of the frame
width = frame.shape[1]

# Print the width
print("Frame width:", width)
```

```python
overlay = frame.copy()
alpha = 0.6
# cv2.addWeighted(overlay, alpha, frame, 1 - alpha, 0, frame)

team_1 = 0.67
team_2 = 1 - team_1
team_1_color = (255, 0, 0)
team_2_color = (0, 255, 0)

cv2.rectangle(overlay, (width//2-250, 940), (width//2+250, 980), team_2_color, -1)
cv2.putText(
    overlay,
    f"{team_2*100:.2f}%",
    (width//2+150, 970),
    cv2.FONT_HERSHEY_SIMPLEX,
    0.8,
    (0, 0, 0),
    3,
)

team_1_box_coordinates = width//2 - 250 + team_1 * 500
cv2.rectangle(overlay, (width//2-250, 940), (int(team_1_box_coordinates), 980), team_1_color, -1)
cv2.putText(
    overlay,
    f"{team_1*100:.2f}%",
    (width//2-240, 970),
    cv2.FONT_HERSHEY_SIMPLEX,
    0.8,
    (0, 0, 0),
    3,
)
cv2.addWeighted(overlay, alpha, frame, 1 - alpha, 0, frame)

plt.imshow(cv2.cvtColor(overlay, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB
plt.axis('off')
plt.show()
```
